```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dat1a=pd.read_excel('Assignment-1_Data.xlsx')
dat1a.head()
```

```python
 
dat1a['Itemname']=dat1a['Itemname'].str.strip()
dat1a.dropna(axis=0,subset=['BillNo'],inplace=True)
dat1a['BillNo']=dat1a['BillNo'].astype('str')
dat1a=dat1a[~dat1a['BillNo'].str.contains('C')]
```

```python
dat1a['Country'].value_counts()
```

```python
dat1a['Country'].shape
```

```python
data=(dat1a[dat1a['Country']=="Germany"]
      .groupby(['BillNo','Itemname'])['Quantity']
      .sum().unstack().reset_index().fillna(0)
      .set_index('BillNo'))
```

```python
data.head()
```

```python
def my_encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1

my_basket_sets=data.applymap(my_encode_units)
my_basket_sets.drop('POSTAGE',inplace=True,axis=1)
```

```python
mfi=apriori(my_basket_sets,min_support=0.07,use_colnames=True)
```

```python
my_rules=association_rules(mfi,metric="lift",min_threshold=1)
```
