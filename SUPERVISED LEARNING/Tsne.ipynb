```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data=pd.read_csv('mall_customers.csv')
```

```python
data.head()
```

```python
data['age'].unique()
data.info()
```

```python
features=['age','annual_income','customer_id']
target=['spending_score']
```

```python
data_norm=data.copy
sc=StandardScaler()
```

```python
sc.fit_transform(data[features])
```

```python
tsne=TSNE(learning_rate=500,n_components=2)
x_tsne = tsne.fit_transform(data[features])
```

```python
y=data[target]
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x_tsne[np.where(y==0),0],x_tsne[np.where(y==0),1],color='g',marker='*')
plt.scatter(x_tsne[np.where(y==1),0],x_tsne[np.where(y==1),1],color='r',marker='+')
```
